```python
import pandas as pd
import numpy as np

data = pd.read_csv("practice.csv")
data.head(5)
```

```python
# change columns
data.columns = ["timestamp", "gender", "age", "hasBought", 
                "when_both", "process_both", "similar_both",
                "when", "process", "similar",
                "process_pill", "similar_pill"]

# data cleanup
data.when = data.when.fillna(data.when_both)
data.process = data.process.fillna(data.process_both)
data.similar = data.similar.fillna(data.similar_both)

data = data.drop(['when_both', 'process_both', 'similar_both'], axis=1)

data.when = data.when.fillna("")
data.process = data.process.fillna("")
data.similar = data.similar.fillna("")
data.process_pill = data.process_pill.fillna("")

data
```

```python
data.head()
```

### gender - hasBought

```python
# one hot coding
options = data.hasBought.unique()
data["pills"] = data.hasBought.apply(lambda e: options[0] in e or options[3] in e)
data["condom"] = data.hasBought.apply(lambda e: options[1] in e or options[3] in e)

# divide by gender
female = data[data.gender.apply(lambda e: "여성" in e)]
male = data[data.gender.apply(lambda e: "남성" in e)]
```

```python
f_condom = female[female.condom == True] 
m_condom = male[male.condom == True]
print( len(m_condom) / (len(f_condom) + len(m_condom)) )

print(len(f_condom) + len(m_condom))
```

```python
# t test
from scipy.stats import ttest_ind
ttest_ind(female['condom'], male['condom'], equal_var=False)
```

### when

```python
# condom
occasions = data.when.unique()

f_condom_when = []
m_condom_when = []

for o in occasions:
    f_condom_when.append( f_condom[f_condom.when.apply(lambda e: o in e)] )
    m_condom_when.append( m_condom[m_condom.when.apply(lambda e: o in e)] )
    

for i in range(len(occasions)):
    if (i == 0): continue;
    freq = len(f_condom_when[i])
    print(freq, "\t", freq/len(f_condom), "\t", occasions[i])
    
print()
for i in range(len(occasions)):
    if (i == 0): continue;
    freq = len(m_condom_when[i])
    print(freq, "\t", freq/len(m_condom), "\t", occasions[i])

print()
print(len(f_condom))
print(len(m_condom))
```
